## Necessary Library For Reading and Analysing Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from math import sqrt 
```

```python
# loading the data from csv file to a Pandas DataFrame
data = pd.read_csv(train_csv)
```

## Data Preprocessing

```python
X = data.drop(columns=['name', 'status'], axis=0)
Y = data['status']
```

```python
X
```

```python
Y
```

## Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print('X :', X.shape)
print('X_train :', X_train.shape)
print('X_test :', X_test.shape)
```

## Data Standardization

```python
Scaler = StandardScaler()
```

```python
Scaler.fit(X_train)
```

```python
X_train = Scaler.transform(X_train)
X_test = Scaler.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Model Building

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train, Y_train)
```

## Model Evaluation

```python
X_train_prediction = model.predict(X_train)
X_train_prediction
```

```python
# Accuracy Score On Training Data
training_data_score = accuracy_score(Y_train, X_train_prediction)
print('Training Data Score :', training_data_score)
```

```python
X_test_prediction = model.predict(X_test)
X_test_prediction
```

```python
# Accuracy Score On Test Data
test_data_score = accuracy_score(Y_test, X_test_prediction)
print('Test Data Score :', test_data_score)
```

```python
# Errors On Training Data
mse = mean_squared_error(Y_train, X_train_prediction)
print('Mean Squared Error :', mse)

rmse = np.sqrt(mean_squared_error(Y_train, X_train_prediction))
print('Root Mean Squared Error :', rmse)
```

```python
# Errors on Test Data
mse = mean_squared_error(Y_test, X_test_prediction)
print('Mean Squared Error :', mse)

rmse = np.sqrt(mean_squared_error(Y_test, X_test_prediction))
print('Root Mean Squared Error :', rmse)
```
